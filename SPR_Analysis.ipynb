```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 포인트 클라우드 갯수
sizes = [50686, 243179, 805842]
# PointDreamer yaml, SPR
yaml = ['default', 'geo_by_SPR', 'nearest', 'noisy', 'wo_NBF', 'SPR']
# 실행 시간
run = [
    [263, 224, 84, 256, 259, 14],
    [294, 266, 113, 337, 293, 60],
    [544, 454, 384, 647, 548, 216]
]

# 시각화
plt.figure(figsize=(10, 6))

for i, size in enumerate(sizes):
    plt.plot(yaml, run[i], marker='o', label=str(size))  # 숫자만 라벨로 사용

plt.title("포인트 클라우드 수와 yaml 설정에 따른 실행 시간")
plt.xlabel("yaml 설정")
plt.ylabel("실행 시간 (초)")
plt.legend(title="Point size")  # 범례에 제목만 추가 가능
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 포인트 클라우드 갯수
sizes = [100000, 200000, 400000, 800000]
# SPR depth
depth = [7, 8, 9, 10, 11, 12]
# 생성된 포인트 클라우드 갯수
generates = [
    [24750, 90283, 310131, 569146, 676937, 695107],
    [25268, 95885, 328527, 737407, 1370227, 1546006],
    [25238, 103847, 386158, 1295531, 2365571, 2641909],
    [24922, 103328, 393775, 1310154, 2553207, 3110549]
]

# 시각화
plt.figure(figsize=(10, 6))

for i, generate in enumerate(generates):
    plt.plot(depth, [j/generate for j in run[i]], marker='o', label=str(size)+'개')  # 숫자만 라벨로 사용

plt.title("SPR depth에 따른 포인트 클라우드 수 증가비")
plt.xlabel("SPR depth")
plt.ylabel("증가비")
plt.legend(title="Point size")
plt.grid(True)
plt.tight_layout()
plt.show()
```
