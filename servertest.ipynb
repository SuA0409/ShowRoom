```python
import viser
import torch
import numpy as np
import cv2
from pyngrok import ngrok, conf
import socket
import warnings
import glob
import os

def find_free_port():
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind(('', 0))  # 운영체제가 자동으로 포트 할당
        return s.getsockname()[1]

# 서버 주소 할당
def make_server(token):
    port = find_free_port() # 빈 포트 추출
    conf.get_default().auth_token = token # Ngrok 설정

    url = ngrok.connect(port, "http") # 새 Ngrok 터널 생성
    server = viser.ViserServer(host="0.0.0.0", port=port) # Viser 서버 실행 (단일 서버)
    return url, server # url, server 반환 (url 고정한 채로 서버는 계속 쓸 수 있음)

url1, server = make_server('2xwLYd6T4TFcrLjwgJZpxbDgaOJ_7oCZnw8f4Bkx2sYX3zkGQ')

print(url1)
```

```python
from flask import Flask, request, jsonify
import torch
import numpy as np
import os
import time
from fast3r.models.multiview_dust3r_module import MultiViewDUSt3RLitModule
import viser
import numpy as np
import threading

@server.on_client_connect
def on_client_connect(client: viser.ClientHandle) -> None:
    with client.atomic():
        client.camera.position = (-0.00141163, -0.01910395, -0.06794288)
        client.camera.look_at = (-0.00352821, -0.01143425, 0.0154939)
    client.flush()
server.scene.set_up_direction((0.0, -1.0, 0.0))
server.scene.world_axes.visible = False

def run_viser():
    server.serve_forever()

# viser 서버를 별도의 쓰레드로 실행
threading.Thread(target=run_viser, daemon=True).start()

@app.route('/viser', methods=['POST'])
def viser_route():
    try:
        data = np.load('/content/drive/MyDrive/Final_Server/Input/Pts/fast3r_output.npz', allow_pickle=True)
        pc = data['point_cloud']
        color = data['color']
        print(pc.shape, color.shape)
        print(f'포인트 클라우드 최대값 :{pc[0].max()}    포인트 클라우드 최소값 : {pc[0].min()}')

        # 이미 실행중인 server에 데이터만 추가
        server.add_point_cloud(
            name="point_cloud",
            points=pc,
            colors=color,
            point_size=0.001
        )
        return jsonify({"status": str(url1)})
    except Exception as e:
        return jsonify({"status": "fail", "error": str(e)})

@app.route('/3d_upload', methods=['POST'])
def echo():
    try:
        print(' 1. 메인에서 요청 받음 !')
        time.sleep(1)

        path = '/content/drive/MyDrive/Final_Server/Input/Images'
        sample = len(os.listdir(path))
        print(f"    받은 이미지의 개수 : {sample}")

        room, color = batch_images_load('/content/drive/MyDrive/Final_Server/Input/Images', batch_size=1, size=512, sample=sample)
        for i, image in enumerate(room):
            room[i]['img'] = image['img'].to(device)
            room[i]['true_shape'] = image['true_shape'].to(device)

        print(f"🔎 [서버] room[0] keys: {list(room[0].keys())}")
        print(f"🔎 [서버] room[0][true_shape]: {room[0]['true_shape']}")

        # 🔎 모델 추론
        print(' 2. model 추론 시작...')
        with torch.no_grad():
            pred = model(room)
        print('    model 추론 완료')

        poses_c2w_batch, _ = MultiViewDUSt3RLitModule.estimate_camera_poses(
            pred,
            niter_PnP=100,
            focal_length_estimation_method='first_view_from_global_head'
        )

        camera_poses = poses_c2w_batch[0]
        with open('/content/drive/MyDrive/Final_Server/Input/Poses/poses.txt', 'w') as f:
            f.write(str(camera_poses))

        # 🔎 결과 저장
        pcs = np.concatenate([np.reshape(pred[i]['pts3d_in_other_view'].cpu().numpy().squeeze(), (-1, 3)) for i in range(sample)], axis=0)
        color = np.concatenate([np.reshape(color[i], (-1, 3)) for i in range(sample)], axis=0)


        print(' 3. 이미지 후처리 완료')

        np.savez('/content/drive/MyDrive/Final_Server/Input/Pts/fast3r_output.npz', point_cloud=pcs, color=color)

        print(' 4. point_could & color 저장 완료!')



        print(' 5. SPR 후처리 완료!')
        return jsonify({"status": "success"})
    except Exception as e :
        return jsonify({"status": str(e)})

app.run(host='0.0.0.0', port=5001)
```
